# Simulation of Ehrenfest Model

```python
import numpy as np
import matplotlib.pyplot as plt
from math import comb

def ehrenfest_urn(N, steps, seed=None):
    rng = np.random.default_rng(seed)
    k = N
    states = np.zeros(steps + 1, dtype=int)
    states[0] = k
    
    for t in range(steps):
        if rng.random() < k / N:
            k -= 1
        else:
            k += 1
        states[t + 1] = k
    
    return states


def plot_deviation(states, N, steps):
    deviation = states - N / 2
    plt.figure()
    plt.plot(deviation)
    plt.xlabel("Time Step")
    plt.ylabel("Deviation from N/2")
    plt.title(f"Deviation Over Time (Steps={steps})")
    plt.show()




if __name__ == "__main__":
    N = 100
    step_values = [100, 1000, 10000]
    for steps in step_values:
        states = ehrenfest_urn(N, steps, seed=42)
        
        plot_deviation(states, N, steps)
```
