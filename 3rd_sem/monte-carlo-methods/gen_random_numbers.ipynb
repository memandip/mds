## Random number generation

## Congruential method

```python
def generate_random_number(seed: int, loop_count: int):
    count = 0
    while count < loop_count:
        ran = (16807*seed) % (2**31 - 1)
        count += 1
        seed = ran
    return ran, count
```

```python
generate_random_number(1, 100)
```

```python
# We have sets of randome numbers (r). r belongs to (0, 1)

# Use congruential method to generate random numbers between 0 and 1
def congruential_random_numbers(seed: int, n: int):
    random_numbers = []
    for _ in range(n):
        mod = (2**31 - 1)
        seed = (16807 * seed) % mod
        random_numbers.append(seed / mod)
    return random_numbers
```

```python
random_nums = congruential_random_numbers(12414, 1000000)
unique_random_nums = set(random_nums)

print(f"Total random numbers generated: {len(random_nums)}")
print(f"Unique random numbers generated: {len(unique_random_nums)}")
```

Calculate mean & variance.

```python
import random
import numpy as np

total_nums = 10000000
congruential_rand_nums = congruential_random_numbers(12414, total_nums)
builtin_rand_nums = [random.random() for _ in range(total_nums)]

mean_congruential = np.mean(congruential_rand_nums)
mean_builtin = np.mean(builtin_rand_nums)

var_congruential = np.var(congruential_rand_nums)
var_builtin = np.var(builtin_rand_nums)

std_congruential = np.std(congruential_rand_nums)
std_builtin = np.std(builtin_rand_nums)

gussian_congruential = np.random.normal(mean_congruential, std_congruential, total_nums)
gussian_builtin = np.random.normal(mean_builtin, std_builtin, total_nums)

print(f"Congruential Method - Mean: {mean_congruential:.4f}, Variance: {var_congruential:.6f}")
print(f"Built-in Random - Mean: {mean_builtin:.4f}, Variance: {var_builtin:.6f}")
print(f"Congruential Method - Std Dev: {std_congruential:.6f}")
print(f"Built-in Random - Std Dev: {std_builtin:.6f}")
plt.hist(gussian_congruential, bins=50, alpha=0.5, label='Congruential Method')
plt.hist(gussian_builtin, bins=50, alpha=0.5, label='Built-in Random')
plt.legend()
plt.show()
```
