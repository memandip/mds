## Random number generation

## Congruential method

```python
def generate_random_number(seed: int, loop_count: int):
    count = 0
    while count < loop_count:
        ran = (16807*seed) % (2**31 - 1)
        count += 1
        seed = ran
    return ran, count
```

```python
generate_random_number(1, 100)
```

```python
# We have sets of randome numbers (r). r belongs to (0, 1)

# Use congruential method to generate random numbers between 0 and 1
def congruential_random_numbers(seed: int, n: int):
    random_numbers = []
    for _ in range(n):
        mod = (2**31 - 1)
        seed = (16807 * seed) % mod
        random_numbers.append(seed / mod)
    return random_numbers
```

```python
random_nums = congruential_random_numbers(1, 10000000)
unique_random_nums = set(random_nums)

print(f"Total random numbers generated: {len(random_nums)}")
print(f"Unique random numbers generated: {len(unique_random_nums)}")
```
