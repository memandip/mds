## Random number generation

## Congruential method

```python
def generate_random_number(seed: int, loop_count: int):
    count = 0
    while count < loop_count:
        ran = (16807*seed) % (2**31 - 1)
        count += 1
        seed = ran
    return ran, count
```

```python
generate_random_number(1, 100)
```
