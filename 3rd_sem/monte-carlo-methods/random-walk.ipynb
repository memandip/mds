## Random Walk

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the parameters
n_steps = 1_000_000  # Number of steps in the random walk

def random_walk(n_steps):
    steps = [
        1 if np.random.random() > 0.5 else -1
        for _ in range(n_steps)
    ]
    
    # Cumulative sum of steps
    positions = np.cumsum(steps)

    # Starting position
    positions = np.insert(positions, 0, 0)

    # time axis
    time_steps = np.arange(n_steps+1)

    return time_steps, positions

plt.figure(figsize=(10, 6))
# Run 3 random walks
n_walks = 3
for i in range(n_walks):
    time_steps, positions = random_walk(n_steps)
    plt.plot(time_steps, positions, label=f'Walk {i+1}')

time_steps, positions = random_walk(n_steps)

plt.plot(positions)

plt.legend()
plt.show()
```
