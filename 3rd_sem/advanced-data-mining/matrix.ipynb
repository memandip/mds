```python
import numpy as np

matrix = np.arange(25).reshape(5, 5) % 2  
matrix
```

```python
import matplotlib.pyplot as plt

plt.imshow(matrix, cmap='gray')
current_count = 0
# Show indexes on each cell
for i in range(matrix.shape[0]):
    for j in range(matrix.shape[1]):
        plt.text(j, i, str((current_count*5)+j), ha='center', va='center', color='red')
    current_count += 1
plt.title('Binary Matrix Visualization')
plt.show()
```

```python
import matplotlib.pyplot as plt

new_matrix = np.zeros((5, 5))  # Create a 5x5 matrix filled with zeros

t=3
# Show indexes on each cell
for i in range(1,4):
    for j in range(1,4):
        new_matrix[i][j] += t
        t+=1

plt.imshow(new_matrix, cmap='Purples')
current_count = 0
# Show indexes on each cell
for i in range(matrix.shape[0]):
    for j in range(matrix.shape[1]):
        plt.text(j, i, str((current_count*5)+j), ha='center', va='center', color='red')
    current_count += 1
```
